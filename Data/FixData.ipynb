extraction de nouveau colonnes depuis la colonne Descriptiion

```python
import csv
import re

# Liste des mots-clés possibles pour chaque champ
KEYWORDS = {
    "Processeur": ["Processeur"],
    "Système d'exploitation": ["Système d'exploitation", "OS"],
    "Mémoire RAM": ["Mémoire RAM", "RAM"],
    "Disque Dur": ["Disque Dur", "SSD", "HDD"],
    "Carte mère": ["Carte mère", "Motherboard"],
    "Carte Graphique": ["Carte Graphique", "GPU", "Graphique"],
    "Dimensions": ["Dimensions", "Taille"],
    "Garantie": ["Garantie"],
    "Couleur": ["Couleur", "Color"]
}

def extract_field(description, keywords):
    """
    Cherche chaque mot clé et retourne la valeur après ':' jusqu'au prochain '-'
    """
    result = {}
    for field, keys in keywords.items():
        pattern = r"(?:{}):\s*(.*?)(?:\s*-\s*|$)".format("|".join(re.escape(k) for k in keys))
        match = re.search(pattern, description, flags=re.IGNORECASE)
        if match:
            result[field] = match.group(1).strip()
        else:
            result[field] = ""
    return result

# Fichiers
input_file = r"C:\Users\azizh\OneDrive\Bureau\TekSpire\Data\DataFix\PCallinone.csv"
output_file = r"C:\Users\azizh\OneDrive\Bureau\TekSpire\Data\DataFix\DataExpand\PcAllinoneEXP.csv"

with open(input_file, newline='', encoding="utf-8") as infile, \
     open(output_file, "w", newline='', encoding="utf-8") as outfile:

    reader = csv.DictReader(infile)
    fieldnames = reader.fieldnames + list(KEYWORDS.keys())
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()

    for row in reader:
        description = row.get("description", "")
        extracted = extract_field(description, KEYWORDS)
        row.update(extracted)
        writer.writerow(row)

print("✅ Extraction intelligente terminée")
```
